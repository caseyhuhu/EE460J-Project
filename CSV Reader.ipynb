```python
import pandas as pd
```

```python
file = open('225171.csv')
rows_to_skip = 0
for line in file:
    if not line.startswith('ball'):
        rows_to_skip += 1
    else:
        break
file.seek(0)

df = pd.read_csv(
    filepath_or_buffer = file,
    skiprows = rows_to_skip,
    names = ["Ball","Innings", "Balls", "Team", "Batsman", "Bowler","Non_Striker", "BatsmanScore", "Extras", "Catch", "Catcher"],
    usecols = range(1,11)
)
over_col = []
ball_col = []
for num in df['Balls']:
    num1, num2 = str(num).split('.')
    over_col.append(num1)
    ball_col.append(num2)
df['Over'] = over_col
df['Ball'] = ball_col
df.drop('Balls', axis=1, inplace=True)
df.head(18)
```
