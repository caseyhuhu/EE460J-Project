```python
import pandas as pd
```

```python
# 1. Find the start of the ball data
filename = '225171.csv'
file = open(filename)
rows_to_skip = 0
for line in file:
    if not line.startswith('ball'):
        rows_to_skip += 1
    else:
        break
file.seek(0)

# 2. Create the DataFrame
df = pd.read_csv(
    filepath_or_buffer = file,
    skiprows = rows_to_skip,
    names = ["Ball","Innings", "Balls", "Bat Team", "Batsman", "Bowler","Non_Striker", "BatsmanScore", "Extras", "Catch", "Catcher"],
    usecols = range(1,11)
)

# 3. Modify columns
over_col = []
ball_col = []
for num in df['Balls']:
    num1, num2 = str(num).split('.')
    over_col.append(num1)
    ball_col.append(num2)
df['Over'] = over_col
df['Ball'] = ball_col
df.drop('Balls', axis=1, inplace=True)
team1 = df['Bat Team'].iloc[0]
team2 = df['Bat Team'].iloc[-1]
df['Bowl Team'] = [team2 if team == team1 else team1 for team in df['Bat Team']] # Python is cool
df['Game ID'] = filename[:-4]

display(df.head(6))
display(df.tail(6))
```
